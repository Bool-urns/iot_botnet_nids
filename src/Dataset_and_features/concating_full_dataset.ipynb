```
from sklearn.preprocessing import LabelEncoder

def preprocess_data(df):
  label_encoder = LabelEncoder()
  
  proto_values = df['proto'].values
  flg_values = df['flgs'].values
  state_values = df['state'].values

  proto_encoded = label_encoder.fit_transform(proto_values)
  flg_encoded = label_encoder.fit_transform(flg_values)
  state_encoded = label_encoder.fit_transform(state_values)

  df['proto_number'] = proto_encoded
  df['flgs_number'] = flg_encoded
  df['state_number'] = state_encoded
  

  df = df.drop(['proto', 'flgs', 'state'], axis=1)

  return df
```

# Applying the label encoding function to make the dataset

```
normal_df = preprocess_data(normal_df)

data_theft_df = preprocess_data(data_theft_df)
ddos_http_df = preprocess_data(ddos_http_df)
dos_http_df = preprocess_data(dos_http_df)
keylog_df = preprocess_data(keylog_df)
os_scan_df = preprocess_data(os_scan_df)

```
